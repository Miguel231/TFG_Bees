```python
#threshold of temperature for thermoregulation
TEMP_MIN = 36
TEMP_MAX = 37

#dataframe of the heart module — only rows with data
df_heart = all_boxes[all_boxes['Temperature heart'].notna()][['Time', 'Temperature heart', 
                                                'Humidity heart', 'Frequency', 
                                                'Volume']].reset_index(drop=True)

df_heart['thermoregulating'] = df_heart['Temperature heart'].between(TEMP_MIN, TEMP_MAX)

#how many
n_termo = df_heart['thermoregulating'].sum()
pct_termo = n_termo / len(df_heart) * 100
print(f"Number of hives: {all_boxes['Hive name'].nunique()} and Years: {all_boxes['year'].nunique()}")
print(f"Percentage in thermoregulation (36-37°C): {n_termo} ({pct_termo:.1f}%)")
```

```python
def analyze_thermo(df_heart, df_scale, label, path):

    df_heart_clean = df_heart.reset_index(drop=True)

    # Detect blocks of continuous thermoregulation
    df_heart_clean['block'] = (
        df_heart_clean['thermoregulating'] != df_heart_clean['thermoregulating'].shift()
    ).cumsum()

    blocks = df_heart_clean[df_heart_clean['thermoregulating']].groupby('block').agg(
        start=('Time', 'min'),
        end=('Time', 'max'),
        duration_hours=('Time', lambda x: (x.max() - x.min()).total_seconds() / 3600),
        avg_temp=('Temperature heart', 'mean'),
        records=('Time', 'count')
    ).reset_index(drop=True)

    large_blocks = blocks[blocks['duration_hours'] >= 6].round(2)

    print(f"\n{label}")
    print(f"Thermoregulation blocks (>6h): {len(large_blocks)}")
    print(large_blocks.to_string())

    # Weight variation during thermoregulation
    weight_variation = []

    for _, row in large_blocks.iterrows():

        mask = (df_scale['Time'] >= row['start']) & (df_scale['Time'] <= row['end'])
        weights = df_scale[mask]['Weight']

        if len(weights) >= 2:
            weight_variation.append({
                'start': row['start'],
                'duration_hours': row['duration_hours'],
                'avg_temp': row['avg_temp'],
                'weight_start': weights.iloc[0],
                'weight_end': weights.iloc[-1],
                'weight_change_kg': round(weights.iloc[-1] - weights.iloc[0], 3),
                'weight_mean': round(weights.mean(), 3)
            })

    df_variation = pd.DataFrame(weight_variation)

    if df_variation.empty:
        print("No weight data available for these blocks")
        return

    print(df_variation.to_string())

    # Plot results
    fig, axes = plt.subplots(1, 2, figsize=(14,5))

    colors = ['tomato' if v < 0 else 'seagreen' for v in df_variation['weight_change_kg']]

    # Weight variation bar plot
    axes[0].bar(range(len(df_variation)), df_variation['weight_change_kg'], color=colors)
    axes[0].axhline(0, linewidth=0.8)
    axes[0].set_xticks(range(len(df_variation)))
    axes[0].set_xticklabels(
        [str(d.date()) for d in df_variation['start']],
        rotation=45,
        ha='right',
        fontsize=8
    )
    axes[0].set_title(f'Weight variation during thermoregulation ({label})')
    axes[0].set_ylabel('Weight change (kg)')

    # Temperature vs weight scatter
    axes[1].scatter(
        df_variation['avg_temp'],
        df_variation['weight_change_kg'],
        c=colors,
        s=df_variation['duration_hours'] * 10,
        alpha=0.8
    )
    axes[1].axhline(0, linewidth=0.8)
    axes[1].set_title('Average temperature vs weight change')
    axes[1].set_xlabel('Average temperature (°C)')
    axes[1].set_ylabel('Weight change (kg)')

    plt.tight_layout()
    plt.show()

    #save the figure
    #fig.savefig(f'{path}/{label}_thermo_weight_variation.png')

def remove_weight_spikes(df, column='Weight', max_change_per_step=3):

    df = df.sort_values('Time').copy()

    diff = df[column].diff().abs()

    df.loc[diff > max_change_per_step, column] = None

    df[column] = df[column].interpolate()

    return df
```

# Daily

```python
def compute_daily_metrics(df_heart, df_scale):

    df_heart = df_heart.copy()
    df_scale = df_scale.copy()

    df_heart['date'] = df_heart['Time'].dt.date
    df_scale['date'] = df_scale['Time'].dt.date

    # temperature stability
    temp_daily = df_heart.groupby('date')['Temperature heart'].agg(
        temp_mean='mean',
        temp_std='std'
    )

    # weight slope
    weight_daily = df_scale.groupby('date')['Weight'].agg(
        weight_start='first',
        weight_end='last'
    )

    weight_daily['weight_change'] = (
        weight_daily['weight_end'] - weight_daily['weight_start']
    )

    daily = temp_daily.join(weight_daily)

    return daily.reset_index()

def plot_colony_state(daily, label):

    fig, ax = plt.subplots(figsize=(8,6))

    scatter = ax.scatter(
        daily['temp_std'],
        daily['weight_change'],
        s=60,
        alpha=0.7
    )

    ax.axhline(0)
    ax.axvline(0.5)

    ax.set_xlabel("Temperature stability (std °C)")
    ax.set_ylabel("Daily weight change (kg)")
    ax.set_title(f"Colony behaviour map ({label})")

    plt.show()
```

```python
path = '../../plots'  # local output folder (gitignored)
def prepare_timeseries(df_heart, df_scale):

    df_heart = df_heart.sort_values('Time').copy()
    df_scale = df_scale.sort_values('Time').copy()

    # rolling temperature stability (brood regulation)
    df_heart['temp_std_3h'] = (
        df_heart['Temperature heart']
        .rolling(window=18, center=True)   # adjust depending on sampling rate
        .std()
    )

    return df_heart, df_scale

def plot_colony_timeline(df_heart, df_scale, label):

    fig, axes = plt.subplots(3, 1, figsize=(14,8), sharex=True)

    # Weight
    axes[0].plot(df_scale['Time'], df_scale['Weight'])
    axes[0].set_ylabel("Weight (kg)")
    axes[0].set_title(f"Colony sensor timeline ({label})")

    # Temperature
    axes[1].plot(df_heart['Time'], df_heart['Temperature heart'])
    axes[1].set_ylabel("Brood temperature (°C)")

    # Temperature stability
    axes[2].plot(df_heart['Time'], df_heart['temp_std_3h'])
    axes[2].axhline(0.5, linestyle='--')  # brood regulation threshold
    axes[2].set_ylabel("Temp stability (std °C)")
    axes[2].set_xlabel("Time")

    plt.tight_layout()
    plt.show()

    #save the figure
    #fig.savefig(f'{path}/{label}_colony_timeline.png')
```
